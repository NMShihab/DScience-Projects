# Import package

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Import  datasets from seaborn

```python
flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
```

```python
# First five data 
tips.head()
```

```python
# first five data
flights.head()
```

## Heat map or Matrix plot

```python
# Correlation of tips dataset
corr_tips = tips.corr()
corr_tips
```

```python
# Visualize correlation
sns.heatmap(corr_tips)
plt.show()
```

```python
# Visualize correlation using annot

sns.heatmap(corr_tips,cmap='coolwarm',annot=True)
```

```python
flights.head()
```

```python
# Rearrage the flight data
flights_p = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
# Visualize heat map
sns.heatmap(flights_p,linecolor='white',linewidths=.5)
```

```python
# Cluster map
sns.clustermap(flights_p,cmap='coolwarm')
```

# Grid plot

```python
# import data
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# just grid
sns.PairGrid(iris)
```

```python
# Lets map the grid
grid = sns.PairGrid(iris)
grid.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
grid = sns.PairGrid(iris)
grid.map_diag(plt.hist)
grid.map_upper(plt.scatter)
grid.map_lower(sns.kdeplot)
```

```python
# Pair plot
sns.pairplot(iris)
```

```python
# pair plot with palette
sns.pairplot(iris,hue = 'species',palette='rainbow')
```

# FacetGrid

```python
fgrid = sns.FacetGrid(tips, col="time",  row="smoker")
fgrid.map(plt.hist,'total_bill')
```

```python
fgrid = sns.FacetGrid(tips, col="time",  row="smoker")
fgrid.map(sns.distplot,'total_bill')
```

```python
fgrid = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Notice hwo the arguments come after plt.scatter call
fgrid = fgrid.map(plt.scatter, "total_bill", "tip").add_legend()
```

# JointGrid


```python
jgrid = sns.JointGrid(x="total_bill", y="tip", data=tips)
jgrid = jgrid.plot(sns.regplot, sns.distplot)
```

# LM plot

```python
# Linear model plot
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',
           markers=['v','*'],scatter_kws={'s':100},height=5,palette='bright')
```

```python
# Grid
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',aspect=0.5,height=6)
```

# Style and Color

```python
# default count plot
sns.countplot(x='sex',data=tips)
```

```python
# Set style whitegrid
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
# Set style darkgrid
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
# Set style white
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

## Spine removal

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
```
