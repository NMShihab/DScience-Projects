```python
import bokeh
from bokeh.io import output_notebook,show,reset_output
from bokeh.plotting import figure

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
output_notebook()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## Line plot

```python
from bokeh.models import HoverTool


# Data
x = np.arange(20)
y = np.random.rand(20)

# Line plot
line_plot = figure(plot_width=500,plot_height=300,title='Line chart',x_axis_label='x',y_axis_label='y')
line_plot.line(x,y,legend='line',line_width=2)

# Add hover tools
line_plot.add_tools(HoverTool())

# show the plot
show(line_plot)
```

```python
value = pd.DataFrame(iris.species.value_counts())
value.head()
```

## Bar chart

```python
output_notebook()
x = value.index.values
y = value.species.values


bar_chart = figure(x_range=x,plot_height=300,title="Flower count")
bar_chart.vbar(x=x,top=y,width=.9)
bar_chart.y_range.start = 0


show(bar_chart)
```
