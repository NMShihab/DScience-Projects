# Import  packages

```python
from bs4 import BeautifulSoup
import requests
import re
```

**Specify the url**

```python
url = "https://en.wikipedia.org/wiki/Lionel_Messi"
```

**Get html content from the url**

```python
html = requests.get(url).text
```

**Show the html for the website**

```python
html
```

**Lets parse the html content using Beautiful soup**

```python
soup = BeautifulSoup(html,"html.parser")
```

**Show the parse data**

```python
soup.prettify
```

**Lets extract the website data one by one like:(title,link,table etc)**

```python
# Title of the website
print("Title of the website : ",soup.head.title.text)
```

**Lionel messi information**

```python
personal_info = soup.find("table",class_='infobox vcard')
```

```python
print("The caption is : ",personal_info.caption.text)
```

```python
personal_info.text
```

```python
personal_info.prettify
```

```python
for td in personal_info.find_all('td'):
    print(td.text)
```

**Write a function which return all the link of the web page**

```python
def get_link(page):
    links = []
    for link in page.findAll('a',attrs={'href': re.compile("^http://")}):
        links.append(link.get('href'))
    return links
```

**get all links of the webpage**

```python
all_links = get_link(soup)
```

**How many links are there?**

```python
len(all_links)
```
